```python
img7 = np.array(plt.imread('./images/Charlie.jpg'))
```

```python
bilateral = cv2.bilateralFilter(img7,9,100,100)
```

```python
plt.figure(figsize=[10,10])
plt.subplot(121);plt.imshow(img7);plt.title("Original Image")
plt.subplot(122);plt.imshow(bilateral[:,:,::-1]);plt.title("After bilateral filter")
```

```python
img8 = np.array(plt.imread('./images/house.jpg'))
bilateral_h = cv2.bilateralFilter(img8,9,75,75)
plt.figure(figsize=[10,10])
plt.subplot(121);plt.imshow(img8);plt.title("Original Image")
plt.subplot(122);plt.imshow(bilateral_h);plt.title("After bilateral filter")
```
